```python
import fastf1 as ff1
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import Normalize

# Load sessions
session_2023 = ff1.get_session(2023, 'Silverstone', 'R')
session_2023.load()

session_2024 = ff1.get_session(2024, 'Silverstone', 'R')
session_2024.load()

# Get telemetry data for Lando Norris ('NOR')
laps_2023 = session_2023.laps.pick_driver('NOR')
laps_2024 = session_2024.laps.pick_driver('NOR')

# Pick the fastest lap for both years and extract telemetry data
fastest_lap_2023 = laps_2023.pick_fastest()
fastest_lap_2024 = laps_2024.pick_fastest()

telemetry_2023 = fastest_lap_2023.get_telemetry()
telemetry_2024 = fastest_lap_2024.get_telemetry()

# Ensure both telemetry data sets have the same length
min_len = min(len(telemetry_2023), len(telemetry_2024))
telemetry_2023 = telemetry_2023.iloc[:min_len]
telemetry_2024 = telemetry_2024.iloc[:min_len]

# Extract X, Y, and speed data from telemetry for both years
x_2023 = telemetry_2023['X'].values
y_2023 = telemetry_2023['Y'].values
speed_2023 = telemetry_2023['Speed'].values

x_2024 = telemetry_2024['X'].values
y_2024 = telemetry_2024['Y'].values
speed_2024 = telemetry_2024['Speed'].values

# Ensure all arrays are of the same length
min_len = min(len(x_2023), len(y_2023), len(speed_2023), len(speed_2024))
x_2023 = x_2023[:min_len]
y_2023 = y_2023[:min_len]
speed_2023 = speed_2023[:min_len]
speed_2024 = speed_2024[:min_len]

# Calculate speed difference between 2024 and 2023
speed_diff = speed_2024 - speed_2023

# Create segments for LineCollection
points = np.array([x_2023, y_2023]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

# Normalize the speed difference
norm = Normalize(vmin=-abs(speed_diff).max(), vmax=abs(speed_diff).max())

# Set up figure and axis for plotting with a black background
fig, ax = plt.subplots(figsize=(10, 10), facecolor='black')
ax.set_facecolor('black')

# Use LineCollection to plot the track with the speed difference heatmap
lc = LineCollection(segments, cmap='seismic', norm=norm, linewidth=4)
lc.set_array(speed_diff[:-1])

# Add the track to the plot
ax.add_collection(lc)

# Adjust the axes to fit the data
ax.autoscale()

# Customize the plot appearance
ax.set_title("Track Heatmap: Speed Differences (2024 vs 2023)", fontsize=14, color='white')
ax.tick_params(colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')

# Remove axis labels and ticks for a cleaner look
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('')
ax.set_ylabel('')

# Add a color bar to show the differences
cbar = fig.colorbar(lc, ax=ax)
cbar.set_label('Speed Difference (km/h)', color='white')
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='white')

plt.show()
```

```python
# Import necessary libraries
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np

# Enable the cache
ff1.Cache.enable_cache('/Users/Jackson/Documents/f1/f1_cache')

# Set plotting styles and suppress future warnings
plotting.setup_mpl(misc_mpl_mods=False, color_scheme=None)

# Define the list of recent races
# We'll get the last 6 races up to the latest race in 2023
current_year = 2024
races_to_analyze = 6

# Get the schedule for the current year
schedule = ff1.get_event_schedule(current_year)

# Filter out any events that haven't occurred yet
completed_events = schedule[schedule['EventDate'] < pd.Timestamp.now()]

# Ensure we have enough races
if len(completed_events) < races_to_analyze:
    races_to_analyze = len(completed_events)

# Get the last completed race index
last_race_index = completed_events.index[-1]

# Get the indexes for the last 6 races
race_indexes = range(last_race_index - races_to_analyze + 1, last_race_index + 1)

# Get the list of events for the last 6 races
events = schedule.loc[race_indexes]

# Drivers to analyze
drivers = ['NOR', 'PIA']  # Norris and Piastri

# Initialize dictionary to hold comparison data
comparison_data = {}

for idx, event in events.iterrows():
    print(f"Processing {event['EventName']}...")

    try:
        # Load the race session
        session = ff1.get_session(event['EventDate'].year, event['EventName'], 'R')  # 'R' for Race
        session.load()

        driver_telemetries = {}

        for driver_code in drivers:
            # Get the driver's laps
            laps = session.laps.pick_driver(driver_code)

            # Check if laps are available
            if laps.empty:
                print(f"No laps found for {driver_code} in {event['EventName']}")
                continue

            # Get the fastest lap for the driver
            fastest_lap = laps.pick_fastest()

            # Get telemetry data for the fastest lap
            telemetry = fastest_lap.get_telemetry()

            # Store the telemetry data
            driver_telemetries[driver_code] = telemetry

        # If we have telemetry data for both drivers, proceed
        if len(driver_telemetries) == 2:
            # Find the overlapping distance range
            min_distance = max(driver_telemetries['NOR']['Distance'].min(), driver_telemetries['PIA']['Distance'].min())
            max_distance = min(driver_telemetries['NOR']['Distance'].max(), driver_telemetries['PIA']['Distance'].max())

            # Create common distance array
            distance_ref = np.linspace(min_distance, max_distance, num=1000)

            # For each driver, interpolate speed, x, y over the common distance
            driver_interp = {}
            for driver_code in drivers:
                telemetry = driver_telemetries[driver_code]

                # Remove duplicate distances
                telemetry = telemetry.drop_duplicates(subset=['Distance'])

                # Interpolate speed, X, Y over common distance
                speed_interp = np.interp(distance_ref, telemetry['Distance'], telemetry['Speed'])
                x_interp = np.interp(distance_ref, telemetry['Distance'], telemetry['X'])
                y_interp = np.interp(distance_ref, telemetry['Distance'], telemetry['Y'])

                driver_interp[driver_code] = {
                    'Speed': speed_interp,
                    'X': x_interp,
                    'Y': y_interp
                }

            # Calculate speed difference
            speed_diff = driver_interp['NOR']['Speed'] - driver_interp['PIA']['Speed']

            # Store the comparison data
            comparison_data[event['EventName']] = {
                'X': driver_interp['NOR']['X'],  # Use Norris's X for plotting
                'Y': driver_interp['NOR']['Y'],  # Use Norris's Y for plotting
                'SpeedDiff': speed_diff
            }
        else:
            print(f"Not enough telemetry data for {event['EventName']}")
            continue

    except Exception as e:
        print(f"An error occurred for {event['EventName']}: {e}")
        continue

# Generate comparison heat maps for each track
for event_name, data in comparison_data.items():
    print(f"Generating comparison heat map for {event_name}")

    X = data['X']
    Y = data['Y']
    SpeedDiff = data['SpeedDiff']

    plt.figure(figsize=(12, 6))

    # Choose a diverging colormap centered at zero
    cmap = mpl.cm.bwr  # Blue-White-Red colormap

    # Normalize the colormap to center at zero
    max_abs_speed_diff = np.max(np.abs(SpeedDiff))
    norm = mpl.colors.Normalize(vmin=-max_abs_speed_diff, vmax=max_abs_speed_diff)

    # Create the scatter plot
    sc = plt.scatter(X, Y, c=SpeedDiff, cmap=cmap, norm=norm, s=5)
    plt.axis('off')
    plt.title(f"{event_name} - Norris vs Piastri Speed Difference")

    # Add a color bar
    cbar = plt.colorbar(sc)
    cbar.set_label('Speed Difference [km/h] (Norris - Piastri)')

    # Save the figure
    plt.savefig(f"{event_name}_Norris_vs_Piastri_comparison.png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

# Example feature data (you'll need to collect actual race data)
data = {
    'qualifying_position': [1, 2, 3, 1, 4],  # Qualifying positions
    'average_speed': [200, 195, 198, 202, 190],  # Average speed during race
    'pit_stops': [2, 2, 3, 1, 2],  # Number of pit stops
    'overtakes': [0, 1, 2, 0, 3],  # Number of overtakes
    'winner': [1, 0, 0, 1, 0]  # 1 if won the race, 0 otherwise
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Define features (X) and target (y)
X = df[['qualifying_position', 'average_speed', 'pit_stops', 'overtakes']]
y = df['winner']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize RandomForest model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')


# Now you can predict for new data (e.g., 2024 Singapore GP data)
```

```python
# Import necessary libraries and enable caching
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Enable the cache
ff1.Cache.enable_cache('/Users/Jackson/Documents/f1/f1_cache')  # Update the path as needed

# Set plotting styles
plotting.setup_mpl(misc_mpl_mods=False, color_scheme=None)
sns.set_style('whitegrid')

# Step 2: Get the list of all completed races in the latest available season
current_year = 2023  # Change to 2024 when data is available

# Get the schedule for the current year
schedule = ff1.get_event_schedule(current_year)

# Filter out any events that haven't occurred yet
completed_events = schedule[schedule['EventDate'] < pd.Timestamp.now()]

# Check if there are completed events
if completed_events.empty:
    print(f"No completed races found for the year {current_year}.")
else:
    # Get the list of events for the completed races
    events = completed_events

    # Teams and drivers to analyze
    # Update the driver codes if necessary
    mclaren_drivers = ['NOR', 'PIA']  # Lando Norris and Oscar Piastri
    ferrari_drivers = ['LEC', 'SAI']  # Charles Leclerc and Carlos Sainz

    # Initialize a DataFrame to collect all data
    all_races_data = pd.DataFrame()

    # Loop over each completed race
    for idx, event in events.iterrows():
        race_name = event['EventName']
        year = event['EventDate'].year

        print(f"Processing {race_name} ({year})...")

        try:
            # Load the race session
            session = ff1.get_session(year, event['EventName'], 'R')  # 'R' for Race
            session.load()

            # Get laps for McLaren and Ferrari drivers
            mclaren_laps = session.laps.pick_drivers(mclaren_drivers)
            ferrari_laps = session.laps.pick_drivers(ferrari_drivers)

            # Available compounds for this race
            compounds = session.laps['Compound'].unique()
            compounds = [compound for compound in compounds if compound in ['SOFT', 'MEDIUM', 'HARD']]

            # Initialize a DataFrame to store lap times and compounds for this race
            laps_data = pd.DataFrame()

            for compound in compounds:
                # Filter laps for each team and compound, excluding in-laps and out-laps
                mclaren_compound_laps = mclaren_laps.pick_tyre(compound).pick_wo_box()
                ferrari_compound_laps = ferrari_laps.pick_tyre(compound).pick_wo_box()

                # Check if laps are available
                if mclaren_compound_laps.empty and ferrari_compound_laps.empty:
                    continue

                # Add team information
                mclaren_compound_laps['Team'] = 'McLaren'
                ferrari_compound_laps['Team'] = 'Ferrari'

                # Combine laps
                combined_laps = pd.concat([mclaren_compound_laps, ferrari_compound_laps])
                combined_laps['Compound'] = compound
                combined_laps['Race'] = race_name  # Add race name

                # Append to the race's DataFrame
                laps_data = pd.concat([laps_data, combined_laps], ignore_index=True)

            # Check if there are laps to analyze for this race
            if laps_data.empty:
                print(f"No lap data available for {race_name}.")
                continue

            # Convert lap times to seconds for easier comparison
            laps_data['LapTimeSeconds'] = laps_data['LapTime'].dt.total_seconds()

            # Append to the all races DataFrame
            all_races_data = pd.concat([all_races_data, laps_data], ignore_index=True)

        except Exception as e:
            print(f"An error occurred while processing {race_name}: {e}")
            continue

    # After collecting data from all races, create bar graphs
    if all_races_data.empty:
        print("No data collected for analysis.")
    else:
        # Group data by Race, Compound, and Team to calculate average lap times
        avg_lap_times = all_races_data.groupby(['Race', 'Compound', 'Team'])['LapTimeSeconds'].mean().reset_index()

        # Pivot the data to have Teams as columns
        pivot_table = avg_lap_times.pivot_table(index=['Race', 'Compound'], columns='Team', values='LapTimeSeconds')
        pivot_table = pivot_table.dropna()  # Remove entries with missing data

        # Reset index to prepare for plotting
        pivot_table.reset_index(inplace=True)

        # Create bar graphs for each compound across all races
        compounds = pivot_table['Compound'].unique()

        for compound in compounds:
            compound_data = pivot_table[pivot_table['Compound'] == compound]

            if compound_data.empty:
                continue

            # Set up the bar graph
            plt.figure(figsize=(12, 6))
            x = np.arange(len(compound_data['Race']))  # Label locations
            width = 0.35  # Width of the bars

            # Plot bars for each team
            plt.bar(x - width/2, compound_data['McLaren'], width, label='McLaren', color='orange')
            plt.bar(x + width/2, compound_data['Ferrari'], width, label='Ferrari', color='red')

            # Add labels, title, and custom x-axis tick labels
            plt.ylabel('Average Lap Time (seconds)')
            plt.title(f'Average Lap Times on {compound} Compound - {current_year} Season')
            plt.xticks(x, compound_data['Race'], rotation=45, ha='right')
            plt.legend()
            plt.tight_layout()

            # Add data labels on top of each bar
            for i in range(len(x)):
                plt.text(x[i] - width/2, compound_data['McLaren'].iloc[i] + 0.1, f"{compound_data['McLaren'].iloc[i]:.2f}", ha='center', va='bottom', fontsize=8)
                plt.text(x[i] + width/2, compound_data['Ferrari'].iloc[i] + 0.1, f"{compound_data['Ferrari'].iloc[i]:.2f}", ha='center', va='bottom', fontsize=8)

            # Save and show the plot
            plt.savefig(f"{current_year}_{compound}_AverageLapTimes_BarGraph.png")
            plt.show()
```

```python
# Import necessary libraries
import fastf1 as ff1
from fastf1 import plotting
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Enable the cache
ff1.Cache.enable_cache('/Users/Jackson/Documents/f1/f1_cache')  # Update the path as needed

# Set plotting styles
plotting.setup_mpl(misc_mpl_mods=False, color_scheme=None)
sns.set_style('whitegrid')

# Seasons to analyze
seasons = [2023]  # Update with the latest season

# Teams and drivers to compare
teams = {
    'McLaren': ['NOR', 'PIA'],         # Lando Norris and Oscar Piastri
    'Red Bull Racing': ['VER', 'PER'], # Max Verstappen and Sergio Perez
    'Mercedes': ['HAM', 'RUS'],        # Lewis Hamilton and George Russell
    'Ferrari': ['LEC', 'SAI']          # Charles Leclerc and Carlos Sainz
}

# Create a mapping from driver abbreviations to teams
driver_team_map = {}
for team, drivers in teams.items():
    for driver in drivers:
        driver_team_map[driver] = team

# Initialize an empty DataFrame to store race results
race_results = pd.DataFrame()

for season in seasons:
    # Get the schedule for the season
    schedule = ff1.get_event_schedule(season)
    
    # Filter out any events that haven't occurred yet
    completed_events = schedule[schedule['EventDate'] < pd.Timestamp.now()]
    
    for idx, event in completed_events.iterrows():
        race_name = event['EventName']
        year = event['EventDate'].year
        round_number = event['RoundNumber']
        
        print(f"Processing {race_name} ({year})...")
        
        try:
            # Load the race session
            session = ff1.get_session(year, event['EventName'], 'R')
            session.load(telemetry=False, weather=False)
            
            # Get race results and make a copy
            results = session.results[['Abbreviation', 'Position', 'Points', 'Status']].copy()
            
            # Add race information
            results['Race'] = race_name
            results['Round'] = round_number
            results['Year'] = year
            
            # Append to the race results DataFrame
            race_results = pd.concat([race_results, results], ignore_index=True)
            
        except Exception as e:
            print(f"An error occurred while processing {race_name}: {e}")
            continue

# Analyze and Visualize Race Finishing Positions

# Filter results for the teams we're interested in
teams_drivers = [driver for drivers in teams.values() for driver in drivers]
filtered_results = race_results[race_results['Abbreviation'].isin(teams_drivers)].copy()

# Map drivers to their teams
filtered_results['Team'] = filtered_results['Abbreviation'].map(driver_team_map)

# Convert Position to numeric (handling non-finishers)
filtered_results['Position'] = pd.to_numeric(filtered_results['Position'], errors='coerce')

# Remove rows with missing positions (e.g., disqualifications or DNFs)
filtered_results = filtered_results.dropna(subset=['Position'])

# Calculate average finishing positions
avg_positions = filtered_results.groupby(['Team', 'Round']).agg({'Position': 'mean'}).reset_index()

# Plot average finishing positions over rounds
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_positions, x='Round', y='Position', hue='Team', marker='o')
plt.gca().invert_yaxis()  # Invert y-axis so that 1st position is at the top
plt.title('Average Finishing Positions Over Rounds')
plt.xlabel('Round Number')
plt.ylabel('Average Finishing Position')
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import fastf1 as ff1
from fastf1 import plotting
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Enable the cache
ff1.Cache.enable_cache('/Users/Jackson/Documents/f1/f1_cache')  # Update the path as needed

# Set plotting styles
plotting.setup_mpl(misc_mpl_mods=False, color_scheme=None)
sns.set_style('whitegrid')

# Seasons to analyze
seasons = [2023]  # Update with the latest season

# Teams and drivers to compare
teams = {
    'McLaren': ['NOR', 'PIA'],         # Lando Norris and Oscar Piastri
    'Red Bull Racing': ['VER', 'PER'], # Max Verstappen and Sergio Perez
    'Mercedes': ['HAM', 'RUS'],        # Lewis Hamilton and George Russell
    'Ferrari': ['LEC', 'SAI']          # Charles Leclerc and Carlos Sainz
}

# Create a mapping from driver abbreviations to teams
driver_team_map = {}
for team, drivers in teams.items():
    for driver in drivers:
        driver_team_map[driver] = team

# FIA Points System
points_mapping = {
    1: 25,
    2: 18,
    3: 15,
    4: 12,
    5: 10,
    6: 8,
    7: 6,
    8: 4,
    9: 2,
    10: 1
}

# Initialize an empty DataFrame to store race results
race_results = pd.DataFrame()

for season in seasons:
    # Get the schedule for the season
    schedule = ff1.get_event_schedule(season)

    # Filter out any events that haven't occurred yet
    completed_events = schedule[schedule['EventDate'] < pd.Timestamp.now()]

    for idx, event in completed_events.iterrows():
        race_name = event['EventName']
        year = event['EventDate'].year
        round_number = event['RoundNumber']

        print(f"Processing {race_name} ({year})...")

        try:
            # Load the race session
            session = ff1.get_session(year, event['EventName'], 'R')
            session.load(telemetry=False, weather=False)

            # Get race results and make a copy
            results = session.results[['Abbreviation', 'Position', 'Status']].copy()

            # Add race information
            results['Race'] = race_name
            results['Round'] = round_number
            results['Year'] = year

            # Append to the race results DataFrame
            race_results = pd.concat([race_results, results], ignore_index=True)

        except Exception as e:
            print(f"An error occurred while processing {race_name}: {e}")
            continue


teams_drivers = [driver for drivers in teams.values() for driver in drivers]
filtered_results = race_results[race_results['Abbreviation'].isin(teams_drivers)].copy()

filtered_results['Team'] = filtered_results['Abbreviation'].map(driver_team_map)

filtered_results['Position'] = pd.to_numeric(filtered_results['Position'], errors='coerce')

filtered_results['RacePoints'] = filtered_results['Position'].map(points_mapping).fillna(0)

fastest_lap_points = pd.DataFrame()

for season in seasons:
    for idx, event in completed_events.iterrows():
        race_name = event['EventName']
        year = event['EventDate'].year
        round_number = event['RoundNumber']

        try:
            session = ff1.get_session(year, event['EventName'], 'R')
            session.load()

            laps = session.laps.copy()

            fastest_lap = laps.loc[laps['LapTime'].idxmin()]
            fastest_driver = fastest_lap['Driver']
            fastest_lap_time = fastest_lap['LapTime']

            driver_result = filtered_results[
                (filtered_results['Abbreviation'] == fastest_driver) &
                (filtered_results['Round'] == round_number)
            ]

            if not driver_result.empty and driver_result.iloc[0]['Position'] <= 10:
                # Add the fastest lap point
                filtered_results.loc[
                    (filtered_results['Abbreviation'] == fastest_driver) &
                    (filtered_results['Round'] == round_number),
                    'RacePoints'
                ] += 1

        except Exception as e:
            print(f"An error occurred while processing fastest lap for {race_name}: {e}")
            continue

# Remove any rows with missing RacePoints
filtered_results = filtered_results.dropna(subset=['RacePoints'])

team_points = filtered_results.groupby(['Team', 'Round']).agg({'RacePoints': 'sum'}).reset_index()

team_points = team_points.sort_values(by=['Team', 'Round'])

team_points['CumulativePoints'] = team_points.groupby('Team')['RacePoints'].cumsum()

plt.figure(figsize=(12, 6))
sns.lineplot(data=team_points, x='Round', y='CumulativePoints', hue='Team', marker='o')
plt.title('Cumulative Points Over Rounds')
plt.xlabel('Round Number')
plt.ylabel('Cumulative Points')
plt.tight_layout()
plt.show()
```

```python
# Analyze Fastest Lap Times

# Collect fastest lap times from each race
fastest_laps = pd.DataFrame()

for season in seasons:
    # Get the schedule for the season
    schedule = ff1.get_event_schedule(season)
    
    # Filter out any events that haven't occurred yet
    completed_events = schedule[schedule['EventDate'] < pd.Timestamp.now()]
    
    for idx, event in completed_events.iterrows():
        race_name = event['EventName']
        year = event['EventDate'].year
        round_number = event['RoundNumber']
        
        print(f"Processing Fastest Laps for {race_name} ({year})...")
        
        try:
            # Load the race session
            session = ff1.get_session(year, event['EventName'], 'R')
            session.load()
            
            # Get laps data
            laps = session.laps
            
            # Map drivers to their teams
            laps['Team'] = laps['Driver'].map(driver_team_map)
            
            # Filter laps for teams we're interested in
            laps = laps[laps['Team'].isin(teams.keys())].copy()
            
            # Convert LapTime to total seconds
            laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()
            
            # Get the fastest lap for each driver
            fastest_driver_laps = laps.groupby(['Driver', 'Team']).apply(lambda x: x.nsmallest(1, 'LapTimeSeconds')).reset_index(drop=True)
            
            # Add race information
            fastest_driver_laps['Race'] = race_name
            fastest_driver_laps['Round'] = round_number
            fastest_driver_laps['Year'] = year
            
            # Append to the fastest laps DataFrame
            fastest_laps = pd.concat([fastest_laps, fastest_driver_laps], ignore_index=True)
            
        except Exception as e:
            print(f"An error occurred while processing fastest laps for {race_name}: {e}")
            continue

# Calculate average fastest lap times per team per round
avg_fastest_laps = fastest_laps.groupby(['Team', 'Round']).agg({'LapTimeSeconds': 'mean'}).reset_index()

# Plot average fastest lap times over rounds
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_fastest_laps, x='Round', y='LapTimeSeconds', hue='Team', marker='o')
plt.title('Average Fastest Lap Times Over Rounds')
plt.xlabel('Round Number')
plt.ylabel('Lap Time (seconds)')
plt.tight_layout()
plt.show()
```

```python
# Analyze Podium Finishes

# Identify podium finishes (positions 1 to 3)
filtered_results['Podium'] = filtered_results['Position'].apply(lambda x: 1 if x in [1, 2, 3] else 0)

# Calculate total podiums per team per round
team_podiums = filtered_results.groupby(['Team', 'Round']).agg({'Podium': 'sum'}).reset_index()

# Calculate cumulative podiums
team_podiums['CumulativePodiums'] = team_podiums.groupby('Team')['Podium'].cumsum()

# Plot cumulative podiums over rounds
plt.figure(figsize=(12, 6))
sns.lineplot(data=team_podiums, x='Round', y='CumulativePodiums', hue='Team', marker='o')
plt.title('Cumulative Podiums Over Rounds')
plt.xlabel('Round Number')
plt.ylabel('Cumulative Podiums')
plt.tight_layout()
plt.show()
```
